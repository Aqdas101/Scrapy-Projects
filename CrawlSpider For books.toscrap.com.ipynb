```python
import scrapy
from scrapy.linkextractors import LinkExtractor
from scrapy.spiders import CrawlSpider, Rule
from scrapy.crawler import CrawlerProcess
```

```python
class book_spider(CrawlSpider):
    name = 'crawl-spider'

    allowed_domains = ['books.toscrape.com']
    start_urls = ['http://books.toscrape.com/catalogue/category/books/fantasy_19/index.html']
    
    # Rule one
    links_books = LinkExtractor(restrict_css= 'h3 > a')
    links_books_rule = Rule(links_books, callback='parse_item', follow=False)

    # Rule two
    link_next = LinkExtractor(restrict_css = '.next > a')
    rule_next = Rule(link_next, follow=True)

    
    #Rule Three
    link_cats = LinkExtractor(restrict_css = '.side_categories > ul > li > ul > li > a')
    rule_cats = Rule(link_cats, follow=True)
    
    rules = (
        links_books_rule,
        rule_next,
        rule_cats,
    )

    def parse_item(self, response):
       yield {
           'book_name' : response.xpath('//h1/text()').extract(),
           'book_price' : response.css('div.product_main > p.price_color::text').get(),
           'book_url' : response.url
       }
```

```python
process = CrawlerProcess({
    'FEED_FORMAT' : 'csv',
    'FEED_URI' : 'books_scrap.csv',
    'CONCURRENT_REQUESTS' : 20,
    'CONCURRENT_REQUESTS_PER_DOMAIN' : 20
})
process.crawl(book_spider)
process.start()
```
